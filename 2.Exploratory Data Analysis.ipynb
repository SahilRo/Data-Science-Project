```python
import pandas as pd
data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data.head()
```

```python
data_clean = pd.read_pickle('data_clean.pkl')
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```

```python
for interviewer, top_words in top_dict.items():
    print(interviewer)
    print(', '.join([word for word, count in top_words[0:14]]))
    print('---')
```

```python
from collections import Counter
words = []
for interviewer in data.columns:
    top = [word for (word, count) in top_dict[interviewer]]
    for t in top:
        words.append(t)
        
words
```

```python
Counter(words).most_common()
```

```python
add_stop_words = [word for word, count in Counter(words).most_common() if count > 4]
add_stop_words
```

```python
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)
```

```python
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index
```

```python
import pickle
pickle.dump(cv, open("cv_stop.pkl", "wb"))
data_stop.to_pickle("dtm_stop.pkl")
```

```python
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16, 6]
for index, interviewer in enumerate(data.columns):
    wc.generate(data_clean.transcript[interviewer])
    
    plt.subplot(3, 4, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    
plt.show()
```

```python
fullname=['a','b','c','d','e','f']
```

```python
unique_list = []
for interviewer in data.columns:
    uniques = data[interviewer].to_numpy().nonzero()[0].size
    unique_list.append(uniques)
data_words = pd.DataFrame(list(zip(fullname,unique_list)), columns=['interviewer', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
d1=data_words
```

```python
total_list = []
for interviewer in data.columns:
    totals = sum(data[interviewer])
    total_list.append(totals)
```

```python
run_times = [60, 59, 80, 60, 67, 73]
data_words['total_words'] = total_list
data_words['run_times'] = run_times
data_words['words_per_minute'] = data_words['total_words'] / data_words['run_times']
data_wpm_sort = data_words.sort_values(by='words_per_minute')
data_wpm_sort
```

```python
import numpy as np

y_pos = np.arange(len(data_words))

plt.subplot(1, 2, 1)
plt.barh(y_pos, data_unique_sort.unique_words, align='center')
plt.yticks(y_pos, data_unique_sort.interviewer)
plt.title('Number of Unique Words', fontsize=20)

plt.subplot(1, 2, 2)
plt.barh(y_pos, data_wpm_sort.words_per_minute, align='center')
plt.yticks(y_pos, data_wpm_sort.interviewer)
plt.title('Number of Words Per Minute', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
data_bad_words = data.transpose()[['critic', 'criticism','critics']]
data_profanity = pd.concat([data_bad_words.critic + data_bad_words.criticism,data_bad_words.critics], axis=1)
data_profanity.columns = ['f_word', 's_word']
data_profanity
```

```python
plt.rcParams['figure.figsize'] = [10, 8]

for i, interviewer in enumerate(data_profanity.index):
    x = data_profanity.f_word.loc[interviewer]
    y = data_profanity.s_word.loc[interviewer]
    plt.scatter(x, y, color='blue')
    plt.text(x+1.5, y+0.5, fullname[i], fontsize=10)
    plt.xlim(-5, 155) 
    
plt.xlabel('Number of F Bombs', fontsize=15)
plt.ylabel('Number of S Words', fontsize=15)

plt.show()
```

```python
data_words = data.transpose()[['work', 'like','film']]
data_prof = pd.concat([data_words.work + data_words.like,data_words.film], axis=1)
data_prof.columns = ['AA', 'BB']
data_prof
```
